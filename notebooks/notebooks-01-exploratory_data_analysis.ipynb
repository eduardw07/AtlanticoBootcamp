# Descrição dos dados

## Quantidade de Instâncias: 
O dataset contém informações sobre o consumo de bebidas alcoólicas em diferentes países, onde cada linha representa um país.

## Variáveis e Tipos: 
O conjunto de dados possui cinco variáveis: 
1. country: Qualitativa Nominal
2. beer_servings: Quantitativa Discreta
3. spirit_servings: Quantitativa Discreta
4. wine_servings: Quantitativa Discreta
5. total_litres_of_pure_alcohol: Quantitativa Contínua

## Bibliotecas utilizadas:
pandas

seaborn

## Quantidade de Valores Faltantes:

```python
import pandas as pd

df = pd.read_csv("Downloads/drinks.csv")

valores_faltantes = df.isnull().sum()
print(valores_faltantes)
```

## Perguntas de Partida e Hipóteses
### Perguntas
1. Qual é o país com o maior consumo de cerveja?
2. Existe uma correlação entre os consumos?
3. Quais são os países com o maior e o menor consumo total de álcool?

### Hipóteses
1. Países com um maior consumo de cerveja podem ter uma menor taxa de consumo de vinho.
2. Países com uma alta ingestão de destilados podem ter uma maior taxa de consumo total de álcool.

### Insights
1. O país com o maior consumo de cerveja é a Namibia.

```python
maior_consumo_cerveja = df.sort_values(by='beer_servings', ascending=False)

pais_maior_consumo_cerveja = maior_consumo_cerveja.iloc[0]['country']

print("O país com o maior consumo de cerveja é:", pais_maior_consumo_cerveja)
```

```python
import seaborn as sns

colunas_interesse = df.drop(columns=["country"])
correlacao = colunas_interesse.corr()
sns.heatmap(correlacao,annot=True)

```

2. Os valores de correlação entre total_litres_of_pure_alcohol e beer_servings são altos, o que indica que o aumento de um influencia no consumo do outro positivamente.
Outra analise é que spirit_servings e wine_servings possuem uma correlação fraca.

3. O país com o maior consumo total de álcool é a Andorra com, 695 litros, enquanto o país com o menor consumo é o Afeganistão com um total de 0.

```python
df['total_alcohol'] = df['beer_servings'] + df['spirit_servings'] + df['wine_servings']

pais_maior_consumo_total = df.loc[df['total_alcohol'].idxmax()]['country']
maior_consumo_total = df['total_alcohol'].max()

pais_menor_consumo_total = df.loc[df['total_alcohol'].idxmin()]['country']
menor_consumo_total = df['total_alcohol'].min()

print("O país com o maior consumo total de álcool é:", pais_maior_consumo_total, "com um consumo total de", maior_consumo_total, "litros.")
print("O país com o menor consumo total de álcool é:", pais_menor_consumo_total, "com um consumo total de", menor_consumo_total, "litros.")
```

 
### Informações Adicionais
1. Há uma grande variação nos padrões de consumo de bebidas alcoólicas entre os diferentes países.
2. Alguns países têm um consumo predominante de uma determinada bebida, enquanto outros têm um consumo mais equilibrado entre cerveja, destilados e vinho.
